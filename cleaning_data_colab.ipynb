```python
rig_data_df = pd.read_csv('gdrive/My Drive/Project 4/binary_rig_state_data.csv')
rig_data_df.head()
```

```python
# Eliminating columns that are not necessary for the analysis
rig_data_cleaned_df = rig_data_df.drop(columns=['Diff Press', 'Pump SPM 1', 'Pump SPM 2','Mud Weight In'])
rig_data_cleaned_df
```

```python
# Eliminating all negative values, which are not physically possible

rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df['Hole Depth'] >= 0]

rig_data_cleaned_df
```

```python
# Eliminating the rows from columns Rig Activity, Rig Activity Code, Rig Activity Engine, Rig Activity SubCode that are decimals
# as it is a miscalculation and are spurious data

rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df['Rig Activity SubCode'] == rig_data_cleaned_df['Rig Activity SubCode']//1]

rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df['Rig Activity Engine'] == rig_data_cleaned_df['Rig Activity Engine']//1]

rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df['Rig Activity Code'] == rig_data_cleaned_df['Rig Activity Code']//1]

rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df['Rig Activity'] == rig_data_cleaned_df['Rig Activity']//1]

rig_data_cleaned_df

```

```python
# Finding null values, represented by -999.25
rig_data_cleaned_df.min()
```

```python
# Eliminating rows with null values (i.e -999.25), also eliminating negative Block Height as it is physically not possible and likely an error in calibration

rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["ROP - Average"]!=-999.25]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Pump SPM - Total"]!=-999.25]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Flow In Rate"]!=-999.25]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Block Height"]>=0]
```

```python
# Verifying that all null values are eliminated
rig_data_cleaned_df.min()
```

```python
#Finding the unique Rig Activity
rig_data_cleaned_df["Rig Activity"].value_counts()
```

```python
#Finding the unique Rig Activity Code
rig_data_cleaned_df["Rig Activity Code"].value_counts()
```

```python
#Finding the unique Rig Activity Engine
rig_data_cleaned_df["Rig Activity Engine"].value_counts()
```

```python
#Finding the unique Rig Activity Subcode
rig_data_cleaned_df["Rig Activity SubCode"].value_counts()
```

```python
# Eliminating the Rig Activity Codes that do not make sense based on the definition provided by IADC
# International Association of Drilling Contractors (likely are a calculation error)

rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 97]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 99]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 98]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 59]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 30]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 69]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 40]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 55]
rig_data_cleaned_df = rig_data_cleaned_df[rig_data_cleaned_df["Rig Activity Code"] != 4]

rig_data_cleaned_df["Rig Activity Code"].value_counts()
```
